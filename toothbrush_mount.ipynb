# Making a base/holder for the toothbrush quartet! 

* This is also a test to see how well cadquery & jupyter lab work for my use cases.
* Units are in mm

```python
from functools import reduce

BOX_H = 3
NBRUSH = 4
BRUSH_W = 28.6
BRUSH_D = 28.85
BRUSH_F = 7.5
BEZEL = 5
BOX_W = NBRUSH*BRUSH_W+2*BEZEL


base = cq.Workplane("XY").box(5*BRUSH_W+2*BEZEL, BRUSH_D+2*BEZEL, BOX_H).translate(cq.Vector(0, 0, BOX_H/2))
hole = cq.Workplane("XY").box(BRUSH_W, BRUSH_D, BOX_H).edges("|Z").fillet(BRUSH_F).translate(cq.Vector(0,0,BOX_H/2))
# hole = cq.Workplane("XY")\
#              .move(0, BRUSH_D/2).hLine(BRUSH_W/2).line(-0.2,-1).hLine(-0.3).close().mirrorY()\
#             .extrude(2*BOX_H).edges("|Z").fillet(0.2)

BRUSH_EXTENTS = (NBRUSH-1)*(BRUSH_W+BEZEL)
holes = reduce(lambda a, b: a.union(b), [hole.translate(cq.Vector(-BRUSH_EXTENTS/2 + (i * (BRUSH_W + BEZEL)), 0, 0)) for i in range(NBRUSH)])
d = base.cut(holes)

cq.exporters.export(d.faces(">Z"), "./out.svg", opt={"showAxes": False, "projectionDir": (0, 0, 1)})
print("Exported")
```
